```python
import pandas as pd
import numpy as np

import plotly.graph_objects as go
from alpha_vantage.timeseries import TimeSeries
```

```python
api = 'KNR31Y5HCWTVY4Y7'
```

```python
ts = TimeSeries(key=api, output_format='pandas')
```

```python
data1, metadata = ts.get_daily(symbol='DOW',  outputsize = 'full')
data1.tail()
```

```python
data1.describe()
```

```python
figure = go.Figure(data = [go.Candlestick(
    x = data1.index, 
    low= data1['3. low'], 
    high = data1['2. high'], 
    open = data1['1. open'],
    close = data1['4. close'], 
    increasing_line_color = 'green',
    decreasing_line_color = 'red')
     ]
)
figure.show()
```
