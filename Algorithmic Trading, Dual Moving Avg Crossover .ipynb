```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pandas_datareader import data as web
from alpha_vantage.timeseries import TimeSeries
from datetime import datetime 
plt.style.use('fivethirtyeight')
```

```python
api = 'KNR31Y5HCWTVY4Y7'
```

```python
ts =TimeSeries(key=api, output_format = 'pandas')
```

```python
data, metadata = ts.get_daily(symbol='AIG',  outputsize = 'full')
AIG = data['2010-01-01':]
AIG.head()
```

```python
plt.figure(figsize=(18,6))
plt.plot(AIG['4. close'], label='AIG')
plt.title('American International Group, INC Close Price History')
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.legend(loc = 'best')
plt.show()
```

```python
#simple moving average

moving_avg = AIG['4. close'].rolling(window=30).mean()
AVG30 = pd.DataFrame(moving_avg)
AVG30
```

```python
#long term avg 
moving_avg100 = AIG['4. close'].rolling(window=100).mean()
AVG100 = pd.DataFrame(moving_avg100)
AVG100
```

```python
plt.figure(figsize=(18,6))
plt.plot(AIG['4. close'], label='AIG')
plt.plot(AVG30['4. close'], label='AIG 30 day moving avg')
plt.plot(AVG100['4. close'], label='AIG 100 day moving avg')
plt.title('American International Group, INC Close Price History')
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.legend(loc = 'best')
plt.show()
```

```python
new_data = pd.DataFrame()
new_data['AIG'] = AIG['4. close']
new_data['AVG30'] = AVG30['4. close']
new_data['AVG100'] = AVG100['4. close']
new_data
```

```python
#signal to buy/sell

def buy_sell(new_data):
    buy = []
    sell= []
    signal = -1
    
    for i in range (len(new_data)):
        if new_data['AVG30'][i] > new_data['AVG100'][i]:
            if signal != 1:
                buy.append(new_data['AIG'][i])
                sell.append(np.nan)
                signal = 1
            else:
                buy.append(np.nan)
                sell.append(np.nan)
        elif new_data['AVG30'][i] < new_data['AVG100'][i]:
            if signal != 0:
                buy.append(np.nan)
                sell.append(new_data['AIG'][i])
                signal = 0
            else:
                buy.append(np.nan)
                sell.append(np.nan)
        else:
            buy.append(np.nan)
            sell.append(np.nan)
    return(buy,sell)
        
```

```python
buy_sell = buy_sell(new_data)
new_data['buy'] = buy_sell[0]
new_data['sell'] = buy_sell[1]
```

```python
new_data
```

```python
plt.figure(figsize=(18,6))
plt.plot(new_data['AIG'], label='AIG', alpha = .35)
plt.plot(new_data['AVG30'], label='AVG30',  alpha = .35)
plt.plot(new_data['AVG100'], label='AVG100',  alpha = .35)
plt.scatter(new_data.index, new_data['buy'], label = "Buy",  color = 'green')
plt.scatter(new_data.index, new_data['sell'], label = "Sell", color = 'red')
plt.title('American International Group, INC Close Price History with Buy and Sell')
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.legend(loc = 'best')
plt.show()
```
